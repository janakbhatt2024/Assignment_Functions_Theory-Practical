Q6. Explain the concept of generators in Python and how they are defined.

Answer: Generators are a type of iterator that yield values lazily, meaning they compute each value one at a time and donâ€™t store the entire sequence in memory.

Generators are defined using a function with the yield keyword or using a generator expression with parentheses ().

Advantages of generators include memory efficiency, lazy evaluation, and cleaner code for iterating over large datasets or streams of data.

```python
#Example
def count_up_to(limit):
    count = 1
    while count <= limit:
        yield count  # Yield the current count, pause the function
        count += 1
counter = count_up_to(5)
print(next(counter))
print(next(counter))
```

Q9.  Explain the purpose and usage of the `map()` function in Python.

Answer : The map() function in Python is a built-in function that allows you to apply a given function to all the items in an iterable (such as a list, tuple, or any other iterable) and returns a map object (which is an iterator) containing the results.

```python
# Example
numbers = [1, 2, 3, 4]

# Define a function to square a number
def square(x):
    return x * x

# Use map to apply the square function to each element in the list
result = map(square, numbers)

# Convert the result to a list to see the output
print(list(result))  # Output: [1, 4, 9, 16]
```

Q10. What is the difference between map(), reduce(), and filter() functions in Python?

Answer:

map() is for applying a transformation to each element.

reduce() is for performing a cumulative operation (e.g., summing or multiplying all elements).

filter() is for filtering elements based on a condition (e.g., keeping only even numbers).

Each of these functions allows you to work with iterables in a more functional programming style, making the code concise and more readable for certain tasks.

```python
# Ex-1 map function
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)
print(list(result))  # Output: [2, 4, 6, 8]
```

```python
# Ex-2 Reduce Function
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)  # Output: 10 (1 + 2 + 3 + 4)
```

```python
# Ex-3 Filter Function
numbers = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))  # Output: [2, 4]
```

the list.

Answer :

```python
numbers = [1,2,3,4]
def sum_even_numbers(numbers):
    sum = 0
    for num in numbers:
        if num % 2 == 0:
            sum = sum + num
    return sum
sum_even_numbers(numbers)

```

Q4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

Answer:

```python
def check_primes_in_range(start, end):
    for number in range(start, end + 1):
        if number <= 1:
            print(f"{number} is not a prime number")
        else:
            is_prime = True
            for i in range(2, number):
                if number % i == 0:  # If divisible, it's not a prime
                    is_prime = False
                    break
            if is_prime:
                print(f"{number} is a prime number")
            else:
                print(f"{number} is not a prime number")

# Example usage: Check primes in the range 1 to 200
check_primes_in_range(1, 200)
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

Answer :

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  # Total number of terms to generate
        self.index = 0      # To track the current position in the sequence
        self.a, self.b = 0, 1  # First two numbers in the Fibonacci sequence

    def __iter__(self):
        return self  # Return the iterator object itself

    def __next__(self):
        if self.index >= self.terms:
            raise StopIteration  # Stop iteration when the terms limit is reached

        # Return the current Fibonacci number
        fib_number = self.a

        # Update a and b to the next two numbers in the Fibonacci sequence
        self.a, self.b = self.b, self.a + self.b
        self.index += 1

        return fib_number

# Example usage:
fibonacci = FibonacciIterator(10)  # Create an iterator for the first 10 Fibonacci numbers

# Using the iterator with a for loop
for num in fibonacci:
    print(num)
```

Q6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

Answer :

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i  # Yield 2 raised to the power of i

# Example usage:
for power in powers_of_two(5):  # Generate powers of 2 up to 2^5
    print(power)
```

Q7.  Implement a generator function that reads a file line by line and yields each line as a string.

Answer :

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Use .strip() to remove newline characters
```

Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

Answer:

```python
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100, -10]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Using map() to convert the list of Celsius temperatures to Fahrenheit
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Print the converted temperatures
print(fahrenheit_temps)
```

Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

Answer:

```python
# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Given string
input_string = "Hello, World!"

# Using filter() to remove vowels from the string
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Print the result
print(filtered_string)
```

Order_Number = [34587,98762,77226,88112]

Quantity = [4,5,3,3]

Price_Per_Item = [40.95,56.80,32.95,24.99]

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value of the order is smaller than 100,00 â‚¬.

Write a Python program using lambda and map.

Answer:









```python
# Given data
Order_Number = [34587, 98762, 77226, 88112]
Book_Title_and_Author = [{"Learning Python": "Mark Lutz"},
                         {"Programming Python": "Mark Lutz"},
                         {"Head First Python": "Paul Berry"},
                         {"Einfuhrung in Python3": "Bernd Klein"}]
Quantity = [4, 5, 3, 3]
Price_Per_Item = [40.95, 56.80, 32.95, 24.99]

# Using map with a lambda function
result = list(map(lambda order, qty, price: (
                    order,
                    (qty * price + 10) if (qty * price) < 100 else qty * price),
                  Order_Number, Quantity, Price_Per_Item))

# Print the result
print(result)
```
